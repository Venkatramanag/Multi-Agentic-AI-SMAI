```python
import sqlite3
import pandas as pd
import os

# Path to your Northwind database
db_path = r"C:\Users\T8675\Downloads\northwind.db"

# Folder to save CSV files
output_folder = r"C:\Users\T8675\Databases\Northwind\CSV_Files"
os.makedirs(output_folder, exist_ok=True)  # Create folder if it doesn't exist

# Connect to the database
conn = sqlite3.connect(db_path)

# Get list of all tables
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
table_names = tables['name'].tolist()

print("Tables found:", table_names)
```

```python
for table in table_names:
    df = pd.read_sql(f"SELECT * FROM {table};", conn)
    csv_path = os.path.join(output_folder, f"{table}.csv")
    df.to_csv(csv_path, index=False)
    print(f"Exported {table} to {csv_path}")

conn.close()
```
